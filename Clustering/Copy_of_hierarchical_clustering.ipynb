```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters
## Single,Average,Complete Linkage


```python

# Dendrogram for Single Linkage
plt.figure(figsize=(10, 7))
dendrogram_single = sch.dendrogram(sch.linkage(X, method='single'))
plt.title('Dendrogram (Single Linkage)')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()

# Dendrogram for Average Linkage
plt.figure(figsize=(10, 7))
dendrogram_average = sch.dendrogram(sch.linkage(X, method='average'))
plt.title('Dendrogram (Average Linkage)')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()

# Dendrogram for Complete Linkage
plt.figure(figsize=(10, 7))
dendrogram_complete = sch.dendrogram(sch.linkage(X, method='complete'))
plt.title('Dendrogram (Complete Linkage)')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset
## Single,Average,Complete Linkage


```python
from sklearn.cluster import AgglomerativeClustering
# Training the Hierarchical Clustering model on the dataset with Single Linkage
hc_single = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='single')
y_hc_single = hc_single.fit_predict(X)

# Training the Hierarchical Clustering model on the dataset with Average Linkage
hc_average = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='average')
y_hc_average = hc_average.fit_predict(X)

# Training the Hierarchical Clustering model on the dataset with Complete Linkage
hc_complete = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='complete')
y_hc_complete = hc_complete.fit_predict(X)
```

## Visualising the clusters

```python
# Visualising the clusters for Single Linkage
plt.scatter(X[y_hc_single == 0, 0], X[y_hc_single == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_hc_single == 1, 0], X[y_hc_single == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_hc_single == 2, 0], X[y_hc_single == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_hc_single == 3, 0], X[y_hc_single == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_hc_single == 4, 0], X[y_hc_single == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.title('Clusters of customers (Single Linkage)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()

# Visualising the clusters for Average Linkage
plt.scatter(X[y_hc_average == 0, 0], X[y_hc_average == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_hc_average == 1, 0], X[y_hc_average == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_hc_average == 2, 0], X[y_hc_average == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_hc_average == 3, 0], X[y_hc_average == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_hc_average == 4, 0], X[y_hc_average == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.title('Clusters of customers (Average Linkage)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()

# Visualising the clusters for Complete Linkage
plt.scatter(X[y_hc_complete == 0, 0], X[y_hc_complete == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_hc_complete == 1, 0], X[y_hc_complete == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_hc_complete == 2, 0], X[y_hc_complete == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_hc_complete == 3, 0], X[y_hc_complete == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_hc_complete == 4, 0], X[y_hc_complete == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.title('Clusters of customers (Complete Linkage)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
